```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mpg_raw.csv')
df.head()
```

```python
sns.relplot(x='model_year', y='mpg', data=df);
```

```python
sns.relplot(x='horsepower', y='mpg', data=df);
```

```python
sns.relplot(x='model_year', y='mpg', data=df, hue='origin');
```

```python
sns.relplot(x='horsepower', y='mpg', data=df,hue='origin');
```

```python
#we can add some details by using size argument
sns.relplot(x='horsepower', y='mpg', data=df,hue='origin', size='weight');#size='the column you want to be assigned'
```
