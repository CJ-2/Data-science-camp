```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mpg_raw.csv')
df.head()
```

```python
sns.barplot(x='origin',y='horsepower', data=df);#(x axis,y axis, dataFrame)
```

```python
#we can use Standard deviation
sns.barplot(x='origin',y='horsepower', data=df,estimator=np.std);#(x axis,y axis, dataFrame,estimator=any type you want std or mean)
```
