```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mpg_raw.csv')
df.head()
```

```python
# The relationship between mpg and weight
sns.jointplot(x='mpg',y='weight',data=df, kind='scatter')
```

```python
#also the regression
sns.jointplot(x='mpg',y='weight',data=df, kind='reg')
```
