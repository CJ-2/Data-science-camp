```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
#The relationship between more than two variables
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='smoker');
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='time');
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='day');
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex');
```

```python
sns.catplot(data=df,x='time',y='total_bill',hue='sex',kind='bar');
```

```python
plt.figure(figsize=(20,20))
sns.catplot(data=df,x='time',y='total_bill',hue='sex',col='day',kind='bar');
```
