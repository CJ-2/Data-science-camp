```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
#Numerical Data
```

```python
sns.histplot(data=df,x='total_bill')
plt.show()
```

```python
sns.kdeplot(data=df,x='total_bill');
```

```python
sns.displot(data=df, x='total_bill', kind='hist');
```

```python
sns.displot(data=df, x='total_bill', kind='kde');
```

```python
sns.displot(data=df, x='total_bill', col='sex');
```

```python
sns.displot(data=df, x='total_bill',kind='kde', col='sex');
```

```python
sns.displot(data=df, x='total_bill',kde=True);
```

```python
sns.displot(data=df, x='total_bill', col='sex',kde=True);
```
