```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
#Numerical Vs. Numerical
```

```python
sns.lineplot(data=df, x='size',y='tip');
```

```python
sns.scatterplot(data=df,x='size',y='tip');
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip');
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='reg');
```

```python
#Categorical Vs. Categorical
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='day',y='total_bill',hue='sex');
```

```python
#Numerical Vs. Categorical
```

```python
sns.barplot(data=df,x='time',y='tip');
```

```python
sns.boxplot(data=df,x='time',y='tip');
```

```python
sns.swarmplot(data=df,x='time',y='tip');
```

```python
sns.barplot(data=df,x='day',y='tip');
```

```python
sns.boxplot(data=df,x='day',y='tip');
```

```python
sns.swarmplot(data=df,x='day',y='tip');
```
