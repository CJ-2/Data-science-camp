```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tmdb_movies_data.csv')
```

```python
df.head()
```

```python
#Top five best-selling genres
top_genres = df.groupby('genres').sum().sort_values(by='revenue', ascending=False)
top_genres
```

```python
top_genres = top_genres.head()
top_genres
```

```python
top_genres = top_genres['revenue']
```

```python
top_genres
```

```python
plt.pie(top_genres.values, labels=top_genres.index)
plt.show()
```

```python
#some details by autopct
plt.pie(top_genres.values, labels=top_genres.index,autopct='%1.2f%%')
plt.show()
```

```python
#also the explode argument
plt.pie(top_genres.values, labels=top_genres.index, explode=[0.2,0,0,0,0])
plt.show()
```
