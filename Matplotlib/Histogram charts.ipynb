```python
#the hist() function
plt.hist(df['vote_average'])#(the column)
plt.show()#for showing the graph
```

```python
#giving the graph some details
plt.hist(df['vote_average'])
plt.xlabel('vote Average')#title for x label
plt.ylabel('Frequency')#title for y label
plt.title('Distributtion of vote average')#title for the Graph
plt.show()
```

```python
#giving the graph some details
plt.hist(df['vote_average'],facecolor='b',edgecolor='w')#(column,color of column,color of edge)
plt.xlabel('vote Average')#title for x label
plt.ylabel('Frequency')#title for y label
plt.title('Distributtion of vote average')#title for the Graph
plt.show()
```
