```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tmdb_movies_data.csv')
df.head()
```

```python
#The relationship between the revenue and votes
plt.figure(figsize=(10,6))
plt.scatter(df.vote_average,df.revenue)
plt.xlabel('Vote Average')
plt.ylabel('Revenue')
plt.title('The relationship between the revenue and votes')
plt.show()
```
