```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tmdb_movies_data.csv')
```

```python
df.head()
```

```python
df.groupby('original_title').sum().sort_values('revenue', ascending=False)
```

```python
top_revenue = df.groupby('original_title').sum().sort_values('revenue', ascending=False)
top_revenue.head()
```

```python
movies = top_revenue.head().index
movies
```

```python
revenue = top_revenue.head().revenue.values
revenue
```

```python
#Top five best-selling movies
plt.figure(figsize=(10,6))
plt.bar(movies,revenue)
plt.xlabel('movies')
plt.ylabel('revenue')
plt.title('Top five best-selling movies')
plt.show()
```
