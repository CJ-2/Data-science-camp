```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tmdb_movies_data.csv')
df.head()
```

```python
#Number of movies each year
df['release_year'].value_counts(ascending=True)
```

```python
movies_count = pd.DataFrame(df['release_year'].value_counts(ascending=True))
movies_count
```

```python
movies_count = movies_count.reset_index()
movies_count
```

```python
movies_count.columns = ['year','count']
movies_count
```

```python
#we want to sort the values
movies_count = movies_count.sort_values(by='year')
movies_count
```

```python
#now the line chart
plt.plot(movies_count['year'],movies_count['count'])#(X axis,y axis)
plt.show()
```

```python
# changing the colour
plt.plot(movies_count['year'],movies_count['count'],c = '0.6')#(X axis,y axis,colour)
plt.show()
```

```python
#some details
plt.plot(movies_count['year'],movies_count['count'],c = '0.6')#(X axis,y axis,colour)
plt.xlabel('years')#name of x  axis
plt.ylabel('Movies')#name of y axis
plt.title('number of released movies annually')#name of the graph
plt.xlim(1960,2015)#(start,end) for y axis
plt.show()
```

```python
#plt.xtichks function 
plt.plot(movies_count['year'],movies_count['count'],c = '0.6')#(X axis,y axis,colour)
plt.xlabel('years')#name of x  axis
plt.ylabel('Movies')#name of y axis
plt.title('number of released movies annually')#name of the graph
plt.xlim(1960,2015)#(start,end) for y axis
plt.xticks(range(1960,2016,5))#(start,end,gap)
plt.show()
```

```python
#but Something wrong with size
plt.figure(figsize=(10,6))#(width,height)
#plt.xtichks function 
plt.plot(movies_count['year'],movies_count['count'],c = '0.6')#(X axis,y axis,colour)
plt.xlabel('years')#name of x  axis
plt.ylabel('Movies')#name of y axis
plt.title('number of released movies annually')#name of the graph
plt.xlim(1960,2015)#(start,end) for y axis
plt.xticks(range(1960,2016,5))#(start,end,gap)
plt.show()
```

```python
#if we want to add a marker
plt.figure(figsize=(10,6))#(width,height)
#plt.xtichks function 
plt.plot(movies_count['year'],movies_count['count'],c = '0.6',marker = 'o')#(X axis,y axis,colour)
plt.xlabel('years')#name of x  axis
plt.ylabel('Movies')#name of y axis
plt.title('number of released movies annually')#name of the graph
plt.xlim(1960,2015)#(start,end) for y axis
plt.xticks(range(1960,2016,5))#(start,end,gap)
plt.show()
```
